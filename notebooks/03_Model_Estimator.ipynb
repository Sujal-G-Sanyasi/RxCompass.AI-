```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import pandas as pd
import pickle
```

```python
X_train = pd.read_csv('../data/PreProcessed_dataset/X_train_disease.csv')
y_train = pd.read_pickle('../data/PreProcessed_dataset/y_train_disease.pkl')
X_test = pd.read_csv('../data/PreProcessed_dataset/X_test_disease.csv')
y_test = pd.read_pickle('../data/Preprocessed_dataset/y_test_disease.pkl')
```

```python
X_train
```

## Model Estimators

```python
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
xg_clf = XGBClassifier()
```

```python
## Model Training
rf_clf.fit(X_train, y_train)
xg_clf.fit(X_train, y_train)

## Model Prediction
rf_pred = rf_clf.predict(X_test)
xg_pred = xg_clf.predict(X_test)

rf_pred, xg_pred
```

## Model Evaluation

```python
## Confusion matrix for random forest model

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, accuracy_score, mean_squared_error, confusion_matrix, r2_score
confusion_randomf = confusion_matrix(y_test, rf_pred)

plt.figure(figsize=(25,20))
sns.heatmap(data=confusion_randomf, annot=True)
```

```python
## Model Evaluation
print("RandomForestClassifier Report:\n",classification_report(y_test, rf_pred))
```

```python
print("XGBClassifier Report:\n",classification_report(y_test, xg_pred))
```

## Cross Validating both model for possible overfitting

```python
from sklearn.model_selection import cross_val_score
import numpy as np

## RandomForestClassifier Cross Vlidation
cv_score = cross_val_score(rf_clf, X_train, y_train, cv=5, scoring='accuracy')
cv_score.mean()*100
```

```python
## Xgboost Cross Vlidation
cv_score1 = cross_val_score(xg_clf, X_train, y_train, scoring='accuracy', cv=5)
cv_score1.mean()*100
```

## Final Verdict:
- Even after the Cross-val score our average accuracy is estimated as 100% which is valid after the validation process

## Model Selection:
- RandomForestClassifier estimator:
  
  - Since it is simple and works out-of-the-box even with simple fine-tuning and also can be reproducible whereas xgboost might overfit if slight misconfigure especially during smaller dataset.
