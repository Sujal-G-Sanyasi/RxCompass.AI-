## Feature Engineering 
- Basically, Encoding the features using different encoding libraries and  reducing the "Curse of Dimentionality" with PCA.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data/Cleaned_dataset/Disease_Cleaned_train.csv")
df.head(10)
```

```python
## Removing that extra column

df.drop(columns='Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df['prognosis']
```

```python
df.shape
```

## Dependent and Independent Features 

```python
X = df.drop(columns='prognosis', axis=1)
y = df['prognosis']
```

```python
X
```

```python
y
```

## Encoding Feature based upon the Types
1. Fever = Onehot
2. Cough = Onehot
3. Fatigue = Onehot
4. Diffuiculty Breating = Onehot
5. gender = Onehot
6. Blood_Pressure = Ordinal
7. Cholestrol_level = Ordinal

- Output:
- Outcome Variable = Onehot

```python
## Instances for PCA and Encoder

scaler = StandardScaler()
pca = PCA(n_components=0.95)
onehot = OneHotEncoder()
Ordinal = OrdinalEncoder()
label = LabelEncoder()
```

```python
## Splitting data into Train and Test set.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Scaling 

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Principle Component Analysis

```python
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
X_train_pca = pd.DataFrame(X_train_pca)
X_test_pca = pd.DataFrame(X_test_pca)
```

```python
X_train_pca
```

## Encoding 

```python
y_train_encoded = pd.Series(label.fit_transform(y_train), name='Prognosis')
y_test_encoded = pd.Series(label.transform(y_test), name='Prognosis')
      
```

```python
y_train_encoded
```

## Final Pre-Processed dataset


```python
X_train_pca.to_csv("../data/PreProcessed_dataset/X_train_disease.csv", index=False)
y_train_encoded.to_pickle(open("../data/PreProcessed_dataset/y_train_disease.pkl", "wb"))
X_test_pca.to_csv("../data/PreProcessed_dataset/X_test_disease.csv", index=False)
y_test_encoded.to_pickle(open("../data/PreProcessed_dataset/y_test_disease.pkl","wb"))
```
